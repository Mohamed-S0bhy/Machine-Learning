# Importing Libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

# Loading Dataset

```python
data = pd.read_csv('_rank_salary.csv')
```

```python
data.sample(5)
```

```python
data.info()
```

```python
x = data.iloc[:,1]
x
```

```python
y = data.iloc[:,-1]
y
```

```python
plt.plot(x,y,c='r')
plt.xlabel('X',c='g')
plt.ylabel('Y',c='g')
plt.title('Relationship betwwen X and Y',c='g')
```

```python
x = x.to_numpy().reshape(-1,1)
y = y.to_numpy().reshape(-1,1)
```

# First Model  With LinearRegression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
plt.scatter(x,y,c='b')
plt.plot(x,reg.predict(x),c='r')
plt.xlabel('X',c='g')
plt.ylabel('Y',c='g')
plt.title('Linear Regression',c='g')
```

# Second Model  With PolynomialRegression

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree = 4)
```

```python
poly = pf.fit_transform(x,y)
```

```python
reg2=LinearRegression()
reg2.fit(poly,y)
```

```python
plt.scatter(x,y,c='b')
plt.plot(x,reg2.predict(pf.fit_transform(x)),c='r')
plt.xlabel('X',c='g')
plt.ylabel('Y',c='g')
plt.title('Polynomial Regression',c='g')
```
